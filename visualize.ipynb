```python
import pandas as pd
```

```python
def load(fn, column):
    return pd.read_csv(fn, index_col="timestamp", names=["timestamp", column], parse_dates=True)
```

```python
data = load("data/results-read-10k-elapsed-mean.csv", "elapsed (ms)")
```

```python
data.describe(percentiles=[.5, .75, .9, .99])
```
